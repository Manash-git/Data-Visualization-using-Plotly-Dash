## The following graph are for two barmode

1. in 'overlay' mode, bars are top of one another.
2. in 'group' mode, bars are placed beside each other.
3. in 'relative' mode, bars are stacked above (+) or below (-) zero.

```python
# overlay
barchart = px.bar(
    data_frame=df,
    x="year",
    y="convicts",
    color="gender",               
    opacity=0.9,                  
    orientation="v",              
    barmode='overlay',           
                                                              
# ----------------------------------------------------------------------------------------------------------------   
    labels={"convicts":"Convicts in Maharashtra",
    "gender":"Gender"},          
    title='Indian Prison Statistics',
    width=720,                 
    height=350,                   
    template='gridon',           
)

pio.show(barchart)
```

```python
# group
barchart = px.bar(
    data_frame=df,
    x="year",
    y="convicts",
    color="gender",               
    opacity=0.9,                  
    orientation="v",              
    barmode='group',           
                                                              
# ----------------------------------------------------------------------------------------------------------------   
    labels={"convicts":"Convicts in Maharashtra",
    "gender":"Gender"},          
    title='Indian Prison Statistics',
    width=720,                 
    height=350,                   
    template='gridon',           
)

pio.show(barchart)
```

## Statistical error showing

### Calculating error column

```python
#fake margin of error, standard deviation, or 95% confidence interval
df['err_plus'] = df['convicts']/100
df['err_minus'] = df['convicts']/40
df
```

```python
barchart = px.bar(
    data_frame=df,
    x="year",
    y="convicts",
    color="gender",               
    opacity=0.9,                  
    orientation="v",              
    barmode='relative',           
                                                              
# ----------------------------------------------------------------------------------------------------------------   

    error_y="err_plus",         # y-axis error bars are symmetrical or for positive direction
    error_y_minus="err_minus",  # y-axis error bars in the negative direction

# ----------------------------------------------------------------------------------------------------------------    
    labels={"convicts":"Convicts in Maharashtra",
    "gender":"Gender"},          
    title='Indian Prison Statistics',
    width=1000,                 
    height=500,                   
    template='gridon',           
)

pio.show(barchart)
```
